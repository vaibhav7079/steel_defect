```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("train.csv")
test = pd.read_csv('test.csv')
```

### Z_Scratch for Pastry Defect

```python
y=df[['Pastry', 'Z_Scratch', 'K_Scatch', 'Stains','Dirtiness', 'Bumps', 'Other_Faults']]
```

```python
y
```

```python
X = df.drop(columns=['Pastry', 'Z_Scratch', 'K_Scatch', 'Stains', 'Dirtiness', 'Bumps', 'Other_Faults','id'])
```

```python
X.head()
```

```python
from sklearn.preprocessing import RobustScaler
rs=RobustScaler()
```

```python
X=rs.fit_transform(X)
```

```python
X=pd.DataFrame(X,columns=df.drop(columns=['Pastry', 'Z_Scratch', 'K_Scatch', 'Stains', 'Dirtiness', 'Bumps', 'Other_Faults','id']).columns)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(  X, y['K_Scatch'], test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
train_pred=lr.predict(X_train)
test_pred = lr.predict(X_test)
```

```python
 test_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Train accuracy :: ",accuracy_score(y_train,train_pred))
print("Test accuracy :: ",accuracy_score(y_test,test_pred))
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv = cross_val_score(estimator=lr,X=X,y=y['K_Scatch'],cv=5)
```

```python
cv.mean()
```

```python
 for i in y:
        X_train, X_test, y_train, y_test = train_test_split(  X, y[i], test_size=0.33, random_state=42)
        lr.fit(X_train,y_train)
        train_pred=lr.predict_proba(X_train)[:, 1]
        test_pred = lr.predict_proba(X_test)[:, 1]
        
        #print(f"Train accuracy {i} :: ",accuracy_score(y_train,train_pred))
        #print(f"Test accuracy  {i} :: ",accuracy_score(y_test,test_pred))
        #cv = cross_val_score(estimator=lr,X=X,y=y[i],cv=3)
        sub[i]=lr.predict_proba(test_clean)[:, 1]
        sub[i] =sub[i].apply(lambda x:round(x,2))
```
